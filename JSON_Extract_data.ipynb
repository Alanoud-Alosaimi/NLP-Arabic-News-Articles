# Extract News Articles Dataset

### Import  Main libraries 

```python
import pandas as pd
import json
import os
```

### Extract Dataset

```python
files=os.listdir('dataset/')
dfs = []

for file in files:
    with open('dataset/'+file, encoding='UTF-8') as datafile:
          json_data = pd.json_normalize(json.loads(datafile.read()))
    dfs.append(json_data)
data = pd.concat(dfs, sort=False,ignore_index=True)
```
